```python
def findt(A,y0):
    lastIndex= 1
    t = 0
    
    for i in range(len(A)):
        if A[i]==lastIndex:
            if A[i] != A[7]:
                t += 1
        lastIndex = A[i]
    return t

def findT(A,y0):
    T = 0
    firstTime = True
    for i in range(len(A)):
        if A[i] == A[7]:
            if firstTime:
                firstTime = False
                continue
            T += 1
    return T

def findAlphas(A,y0):
    result=[]
    lastIndex= 1
    firstTime = True
    for i in range(len(A)):
        result.append(lastIndex - A[i] )    
        lastIndex = A[i]
#     result = [(x + 0.000001) /y0  for x in result]
    result = [(x) /1  for x in result]
    result2 = []
    
    for x in result:
        if x <=0.00001:
            result2.append(0) 
        elif x <=0.301:
            result2.append(3)    
        elif x <= 0.601:
            result2.append(6) 
        elif x <=0.901:
            result2.append(9)      
        else:
            print('what?')
    
    
    return result2

def findAlphaMax(result):
    alphamax=max(result)
    n=result.count(alphamax)
    return alphamax , n


def findnewt(A,y0):
    result = findAlphas(A,y0)
    max_ = findAlphaMax(result)[0]
    newt = 0
    for x in result:
        if x == max_:
            break
        if x== 0:
             newt += 1
    return newt  

def findPayment(A):
    counter = 0
    lastIndex= 1

    for i in range(len(A)):
        print(lastIndex - A[i])
        if ((lastIndex - A[i])>0):
            counter+=1
        lastIndex = A[i] 
    return counter
```
